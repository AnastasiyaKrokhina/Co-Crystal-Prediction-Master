```python
from rdkit import Chem
from mordred import Calculator, descriptors  #https://github.com/mordred-descriptor/mordred
calc = Calculator(descriptors, ignore_3D=True)
mol = Chem.MolFromSmiles('c1ccccc1')
calc(mol)[:3] # example 
#pd.DataFrame(calc(mol)[:3], columns=calc.descriptors[:3])
```

```python
# Calculate the descriptors for the first dataset (first co-former)
dataset_mol1 = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/cocrystal_design/data/dataset_mol1.csv')
descriptors_mol1 =[]
for mol in dataset_mol1.smiles1.values:
    try:
        descriptors_mol1.append(calc(Chem.MolFromSmiles(mol)))
    except TypeError:
        descriptors_mol1.append('none')
# Save the dataset1 with the descriptors
dataset1 = pd.DataFrame(descriptors_mol1)
dataset1 = pd.DataFrame(dataset1.values, columns=calc.descriptors)
dataset1
```

```python
# Calculate the descriptors for the second dataset (first co-former)
dataset_mol2 = pd.read_csv('dataset_mol2.csv')
descriptors_mol2 =[]
for mol in dataset_mol2.smiles2.values:
    try:
        descriptors_mol2.append(calc(Chem.MolFromSmiles(mol)))
    except TypeError:
        descriptors_mol2.append('none')
# Save dataset2 with the descriptors
dataset2 = pd.DataFrame(descriptors_mol2)
pd.DataFrame(dataset2.values, columns=calc.descriptors).to_csv('dataset2.csv', index=False)
```
